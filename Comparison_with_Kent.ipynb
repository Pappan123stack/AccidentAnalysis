```python
import matplotlib.pyplot as plt
import numpy as np


categories = ['Total Accidents', 'Fatalities', 'Serious Injuries', 'Slight Injuries']
kent = [4494, 63, 787, 3644]
bradford = [833, 9, 139, 685]

# Normalizing the data for spider graph
max_value = max(kent + bradford)
kent_normalized = [x / max_value for x in kent]
bradford_normalized = [x / max_value for x in bradford]

num_vars = len(categories)

# Splitingthe circle into even parts and save the angles
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# Completing the loop for each dataset
kent_normalized += kent_normalized[:1]
bradford_normalized += bradford_normalized[:1]


fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Drawing one axe per variable and add labels
plt.xticks(angles[:-1], categories)

# Drawing ylabels
ax.set_rlabel_position(0)


ax.plot(angles, kent_normalized, linewidth=1, linestyle='solid', label='Kent')
ax.fill(angles, kent_normalized, 'b', alpha=0.1)
ax.plot(angles, bradford_normalized, linewidth=1, linestyle='solid', label='Bradford')
ax.fill(angles, bradford_normalized, 'r', alpha=0.1)

# Adding a legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


categories = ['Total Accidents', 'Fatalities', 'Serious Injuries', 'Slight Injuries']


kent_values = [4494, 63, 787, 3644]
bradford_values = [833, 9, 139, 685]

# Setting the positions and width for the bars
pos = np.arange(len(categories))
bar_width = 0.35

# Creating the bar graph
fig, ax = plt.subplots()
kent_bars = ax.bar(pos - bar_width/2, kent_values, bar_width, label='Kent', color='royalblue')
bradford_bars = ax.bar(pos + bar_width/2, bradford_values, bar_width, label='Bradford', color='seagreen')

# Adding labels, title, and axes ticks
ax.set_ylabel('Counts')
ax.set_title('Road Traffic Accidents in 2019: Kent vs Bradford')
ax.set_xticks(pos)
ax.set_xticklabels(categories)
ax.legend()

# Function to add a label on top of each bar
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


add_labels(kent_bars)
add_labels(bradford_bars)
plt.show()
```
